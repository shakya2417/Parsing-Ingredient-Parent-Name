```
def clean_text(data,col):
  corpus=[]
  for i in range(len(data[col])):
    text=re.sub(r'http\S+', ' ', data[col][i])
    text=text.lower()
    text = re.sub(r'\s*(\d+|[/+*-])',"",text)
    text=re.sub(r'[@#\$%&\*\(\)\<\>\?\!\;\'\\-\_"]',' ',text)
    text= re.sub(r'\W',' ',text)
    text=re.sub(r'\d',' ',text)
    text=re.sub(r'^[a-z]\s',' ',text)
    text=re.sub(r'\s+[a-z]\s+',' ',text)
    text=re.sub(r'\s+[a-z]$',' ',text)
    text=re.sub(r'\s+',' ',text)
    corpus.append(text)
  return corpus
```

```
import re
df['word']=clean_text(df,'Ingredient Name') #cleaning text
```

```
def tokenize():
  words=[]
  for i in range(len(df)):
    words.append(nltk.word_tokenize(df['word'][i]))
  return words
```

```
df['token']=tokenize() #tokenizing data
```

```
from nltk.stem import WordNetLemmatizer 
nltk.download('wordnet')
  
lemmatizer = WordNetLemmatizer()
```

```
#applying word lemmatizer
s2=[]
for i in range(len(df)):
  s1=[]
  for w in df['token'][i]:
    s1.append(lemmatizer.lemmatize(w))
  s3=" ".join(s1) 
  s2.append(s3)

```

```
df['lamm']=s2
```

```
def lamm_clean(data,col):
  corpus=[]
  for i in range(len(data[col])):
    text=re.sub(r'http\S+', ' ', data[col][i])
    text=text.lower()
    text=re.sub(r"leaf","",text)    # removing unnecessary frequently used word
    text = re.sub(r"cup","",text)
    text = re.sub(r"seed","",text)
    text = re.sub(r"chopped","",text)
    text = re.sub(r"powder","",text)
    text = re.sub(r"cut","",text)
    text = re.sub(r"½","",text)
    text = re.sub(r"inch","",text)
    text = re.sub(r"piece","",text)
    text = re.sub(r"red","",text)
    text = re.sub(r"green","",text)
    text=re.sub(r'\s+[a-z]\s+',' ',text) #removing extra spaces from front
    text=re.sub(r'\s+[a-z]$',' ',text) #removing extra spaces from back
    
    corpus.append(text)
  return corpus
```

```
df['lammclean']=lamm_clean(df,'lamm')
```

```
#filling data in output column
df['output']=None  
for i in range(len(df)):

  if len(nltk.word_tokenize(df['lammclean'][i]))==1:
    df['output'][i]=df['lammclean'][i]

  if len(nltk.word_tokenize(df['lammclean'][i]))==2:

    doc = nlp(df['lammclean'][i])
    displacy.render(doc, style='dep',jupyter=False)  #finding of dependencies of words in sentences
    for token in doc:
      if (token.dep_=='ROOT'):
        df['output'][i]=token.text   #extracting information
  if len(nltk.word_tokenize(df['lammclean'][i]))>2:

    doc = nlp(df['lammclean'][i])
    displacy.render(doc, style='dep',jupyter=False)
    for token in doc:
      if token.dep_=='nsubj' or token.dep_=='ROOT':
        df['output'][i]=token.text    #extracting information

      
    
```
